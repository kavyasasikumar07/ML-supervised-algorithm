```python
train.Sex=train.Sex.replace(['male','female'],[0,1])
train.Embarked=train.Embarked.replace(['S','C','Q'],[0,1,2])
```

```python
train
```

```python
final_dataframe= train[['Survived', 'Pclass', 'Sex', 'Age', 'SibSp',
       'Parch', 'Fare','Cabin','Embarked']]
```

```python
cabin_mapping = {cabin: idx for idx, cabin in enumerate(final_dataframe['Cabin'].unique())}
```

```python
final_dataframe.loc[:, 'Cabin'] = final_dataframe['Cabin'].replace(cabin_mapping)
```

```python
final_dataframe
```

```python
train_y=final_dataframe['Survived']
train_x=final_dataframe[['Pclass','Sex','Age','SibSp','Parch','Fare','Cabin','Embarked']]
```

```python
from sklearn.model_selection import train_test_split
train_data,val_data,train_target,val_target=train_test_split(train_x,train_y,train_size=0.8)
```

```python
model=GaussianNB()
model.fit(train_data,train_target)
val_pred=model.predict(val_data)
```

```python
from sklearn.metrics import accuracy_score
print(format(accuracy_score(val_target,val_pred)*100)+"%")
```
