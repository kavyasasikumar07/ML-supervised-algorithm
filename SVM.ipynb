```python
from sklearn.model_selection import train_test_split
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.svm import SVC
from sklearn import svm
import numpy as np
```

```python
titanic=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/train.csv")
```

```python
titanic.head()
```

```python
titanic .drop(['PassengerId','Name','Ticket'],axis=1,inplace=True)
titanic.head()
```

```python
titanic.isnull()
```

```python
titanic.isnull().sum()
```

```python
titanic[titanic.Embarked.isnull()]
```

```python
titanic.drop(['Cabin'],axis=1,inplace=True)
```

```python
titanic['Age']=titanic['Age'].fillna(titanic['Age'].median())
titanic['Embarked']=titanic['Embarked'].fillna('C')
titanic.Sex=titanic.Sex.replace(['male','female'],[0,1])
titanic.Embarked=titanic.Embarked.replace(['S','C','Q'],[0,1,2])
```

```python
titanic.dropna(inplace=True)
```

```python
test=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/test.csv")
```

```python
test.head()
```

```python
test.isnull().sum()
```

```python
test[test.Embarked.isnull()]
```

```python
test['Age']=test['Age'].fillna(test['Age'].median())
test['Embarked']=test['Embarked'].fillna('C')
test.Sex=test.Sex.replace(['male','female'],[0,1])
test.Embarked=test.Embarked.replace(['S','C','Q'],[0,1,2])
```

```python
test.dropna(inplace=True)
```

```python
features=['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']
```

```python
#in linear,gradient descent
#(gamma=c)
C=0.7
clf=svm.SVC(kernel='linear',C=C).fit(titanic[features],titanic['Survived'])
```

```python
clf
```

```python
prediction=clf.predict(test[features])
```
