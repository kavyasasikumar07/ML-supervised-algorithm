```python
dataset=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Position_Salaries.csv")
dataset
```

```python
y=dataset.iloc[:,2].values
y
```

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression() #call linear regression
lin_reg.fit(x,y)
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg.predict(x),color='blue')
plt.title('Linear Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg2=PolynomialFeatures(degree=2) #set degree
x_poly=poly_reg2.fit_transform(x)
lin_reg_2=LinearRegression() #linear regression
lin_reg_2.fit(x_poly,y)
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg_2.predict(poly_reg2.fit_transform(x)),color='blue')
plt.title('Polynomial Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg3=PolynomialFeatures(degree=3)
x_poly=poly_reg3.fit_transform(x)
lin_reg_3=LinearRegression()
lin_reg_3.fit(x_poly,y)
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg_3.predict(poly_reg3.fit_transform(x)),color='blue')

plt.title('Polynomial Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg_2.predict(poly_reg2.fit_transform(x)),color='blue')
plt.plot(x,lin_reg_3.predict(poly_reg3.fit_transform(x)),color='green')
plt.title('Polynomial Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

```

```python
sns.displot(data=data1,x='Y house price of unit area',kde=True,color='Blue')
```

```python
x=data1.drop('Y house price of unit area',axis=1)
x
```

```python
y=data1['Y house price of unit area']
y
```

```python
polynomial_reg=PolynomialFeatures(degree=2,include_bias=False)#we raise degree to assign or set polynomial features
```

```python
poly_feat=polynomial_reg.fit_transform(x)
poly_feat.shape
```

```python
#after linear regression done,split into test and train
x_train,x_test,y_train,y_test=train_test_split(poly_feat,y,test_size=0.3,random_state=101)#30% assigned to test
```

```python
#fit data(means train the polynomial model)
polymodel=LinearRegression()
polymodel.fit(x_train,y_train)
```

```python
#predict and store the model
y_pred=polymodel.predict(x_test)
```

```python
pd.DataFrame({'Y_Test':y_test,'Y_Pred':y_pred,'Residuals':(y_test-y_pred)})
```

```python
poly_MAE=metrics.mean_absolute_error(y_test,y_pred)
poly_MSE=metrics.mean_squared_error(y_test,y_pred)
poly_RMSE=np.sqrt(poly_MSE) #root mean square
```

```python
pd.Series({'Mean_Absolute_Error':poly_MAE,'Mean_Square_Error':poly_MSE,'Root_Mean_Square':poly_RMSE})
```
